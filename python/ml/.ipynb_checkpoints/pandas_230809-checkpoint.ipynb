## 행추출

```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

```python
df.loc[6, "이름":"영어"] = ["Kim", 50, 60]
df
```

- 반복문으로 실행, 하지만 pandas에서는 굳이 그렇게 할 필요 없음

```python
# 국어점수가 90점 이상이면 "A"
# 국어점수가 80점~89점 "B"
# 국어점수가 70점~79점 "C"
# 나머지는 F

for i in range(len(df.index)):
    if df.loc[i, "국어"] >= 90:
        df.loc[i, "점수"] = "A"
    elif 80 <= df.loc[i, "국어"] <= 89:
        df.loc[i, "점수"] = "B"
    elif 70 <= df.loc[i, "국어"] <= 79:
        df.loc[i, "점수"] = "C"
    else:
        df.loc[i, "점수"] = "F"

print(df)
```

- np.where(조건식,참, np.where(조건식,참, (np.where(조건식,참))))
- np.select(다중조건식, 다중결과값, default-->else조건)

```python
condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

## shift 기능
- 정기적으로 들어오는 데이터와의 차이값 구하기 위해 활용

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()
temp_df.head()
```

```python
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value=0).astype(int)
temp_df['diff'] = temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value=0).astype(int)
temp_df['diff2'] = temp_df['price'] - temp_df['shifted_v2']
temp_df.head()
```

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()
temp_df['shifted_v3'] = temp_df['price'].shift(-1, fill_value=0).astype(int)
temp_df['shifted_v4'] = temp_df['price'].shift(-2, fill_value=0).astype(int)
temp_df.tail()
```

### 코로나 발병률과 사망률

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                           labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                            q=5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                           labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                            q=5, precision=0)
covidtotalsonly.head()
```

```python
cross = pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q'])
cross
```

```python
covidtotals.loc[(covidtotalsonly.total_deaths_q == '매우낮음') & (covidtotalsonly.total_cases_q == '중간')]
```
